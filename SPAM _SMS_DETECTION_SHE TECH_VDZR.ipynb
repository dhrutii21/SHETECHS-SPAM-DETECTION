```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import string
```

```python
df=pd.read_csv("spam.csv",encoding="ISO-8859-1")#encoding is for to take the characters as not case sensitive
```

```python
df.head()
```

```python
df.drop(labels=["Unnamed: 2","Unnamed: 3","Unnamed: 4"],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'v1':'spam','v2':'text'},inplace=True)
```

```python
df.head()
```

```python
df.spam=df.spam.apply(lambda s: True if s=='spam' else False)
```

```python
df
```

```python
df.text=df.text.apply(lambda t: t.lower().translate(str.maketrans('','',string.punctuation)))
```

```python
df=df.sample(frac=1)
```

```python
train_df=df.iloc[:int(len(df)*0.7)]
test_df=df.iloc[int(len(df)*0.7):]
```

```python
frac_spam=df.spam.mean()
print(frac_spam)
```

```python
spam_word=' '.join(df[df.spam==True].text).split(' ')
nonspam_words=' '.join(df[df.spam==False].text).split(' ')
```

```python
common_words=set(spam_word).intersection(set(nonspam_words))
```

```python
train_spam_bow=dict()
for w in common_words:
    train_spam_bow[w]=spam_word.count(w)/len(spam_word)
```

```python
train_nonspam_bow=dict()
for w in common_words:
    train_nonspam_bow[w]=nonspam_words.count(w)/len(nonspam_words)
```

```python
train_spam_bow
```

```python
train_nonspam_bow
```

```python
def predict(t,verbose=False):
    valid_words=[w for w in t if w in train_spam_bow]
    spam_prob=[train_spam_bow[w] for w in valid_words]
    nonspam_prob=[train_nonspam_bow[w] for w in valid_words]
    
    if verbose==True:
        df1=pd.DataFrame()
        df1['words']=valid_words
        df1['spam_prob']=spam_prob
        df1['nonspam_prob']=nonspam_prob
        df1['ratio']=[s/n if n>0 else np.inf for s,n in zip(spam_prob,nonspam_prob)]
        print(df1)
        
        
    spam_score=sum(np.log(w) for w in spam_prob)+np.log(frac_spam)
    nonspam_score=sum(np.log(w) for w in nonspam_prob)+np.log(1-frac_spam)
    
    if verbose:
        print('spam score :%s'%spam_score)
        print('non spam score :%s'%nonspam_score)
    
    return spam_score>=nonspam_score
```

```python
predict('are you at class yet'.split(), verbose=True)
```

```python
predict('urgent urgent call call die'.split(), verbose=False)
```

```python
predict('urgent urgent call call die'.split(), verbose=True)
```

```python
predictions = test_df.text.apply(lambda t: predict(t.split()))
```

```python
frac_spam_messages_correctly_detected = np.sum((predictions == True) & (test_df.spam == True)) / np.sum(test_df.spam == True)
print('Fraction Spam Correctly Detected: %s'%frac_spam_messages_correctly_detected)
```

```python
frac_valid_sent_to_spam = np.sum((predictions == True) & (test_df.spam == False)) / np.sum(test_df.spam == False)
print('Fraction Valid Messages Sent to Spam: %s'%frac_valid_sent_to_spam)
```
